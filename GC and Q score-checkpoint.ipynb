### sets up Jupyter Notebook to display plots with the %matplotlib
### inline, displays the plot in the notebook below the cell that creates the plot

```python
%matplotlib inline 
                  
```

```python
import matplotlib.pyplot as plt
```

### df.plot(), plots Samples names on the x-axis against the GC content % on the y-axis. This code produces a line graph by default but omits the some of the sample names.
### plt.xlabel(), labels the x and y axis
### plt.title(), creates a title for the graph


```python
df.plot(x="Sample Name", y=["GC(%) of fq1 (NI)", "GC(%) of fq2 (I)"])
plt.xlabel("Samples")
plt.ylabel("GC-content of clean reads (%)")
plt.title("GC-content of samples pre and post FMDV infection",fontsize = 12)

```

## Since the above code omits some of the sample names, we tried to produce a bar graph
### df.plot(), plots Samples names on the x-axis against the GC content % on the y-axis
### kind="bar", produces a bar graph and shows all the sample names on the x axis
### plt.xlabel(), labels the x and y axis
### plt.title(), creates a title for the graph

```python
df.plot(x="Sample Name", y=["GC(%) of fq1 (NI)", "GC(%) of fq2 (I)"], kind="bar")
plt.xlabel("Samples")
plt.ylabel("GC-content of clean reads (%)")
plt.title("GC-content of samples pre and post FMDV infection", fontsize = 12)
```
